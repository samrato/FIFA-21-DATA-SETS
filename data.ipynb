```python
# Changing Weight to numeric
mask = (df['Weight'].str.endswith('lbs'))
df['Weight'] = df['Weight'].str.strip('kglbs')
df['Weight'] = pd.to_numeric(df['Weight'])
df_lbs = df[mask]
df.loc[mask, 'Weight'] = df_lbs['Weight'] * 0.4535923
```

```python
df['Weight'] = df['Weight'].astype(int)
df['Weight'].unique()
```

```python
# Changing 'Value" to int type
df['Value'] = df['Value'].str.strip('€')
d = {'M': '*1000000', 'K': '*1000'}
s_convert = df['Value'].replace(d, regex=True).map(pd.eval)
df['Value'] = s_convert.astype(int)
```

```python
df['Value']
```

```python
# Changing 'Wage' to int type
df['Wage'] = df['Wage'].str.strip('€')
d = {'M': '*1000000', 'K': '*1000'}
s_convert = df['Wage'].replace(d, regex=True).map(pd.eval)
df['Wage'] = s_convert.astype(int)
```

```python
df['Wage']
```

```python
# Changing 'Release Clause' to int type
df['Release Clause'] = df['Release Clause'].str.strip('€')
d = {'M': '*1000000', 'K': '*1000'}
s_convert = df['Release Clause'].replace(d, regex=True).map(pd.eval)
df['Release Clause'] = s_convert.astype(int)
```

```python
df['Release Clause']
```

```python
# Stripping ★ symbol from 'W/F', 'SM', and 'IR' columns
df['W/F'] = df['W/F'].str.strip("★").astype(int)
df['SM'] = df['SM'].str.strip("★").astype(int)
df['IR'] = df['IR'].str.strip("★").astype(int)
```
